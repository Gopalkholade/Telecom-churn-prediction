```python
no_inter=['internettype', 'avgmonthlygbdownload', 'onlinesecurity',
       'onlinebackup', 'deviceprotectionplan', 'premiumtechsupport',
       'streamingtv', 'streamingmovies', 'streamingmusic', 'unlimiteddata']
```

```python
for i in no_inter:
    if i in obj:
        churn[i]=churn[i].fillna("NA")
    elif i in num:
        churn[i]=churn[i].fillna(0).astype(float)
```

```python
# furthe classify obj into binary and multi-dimentional
uniques={}
nuniques={}
for i in obj:
    uniques[i]=churn[i].unique()
    nuniques[i]=churn[i].nunique()
```

```python
ord_c=[]
ohe_c=[]
for i in nuniques:
    if nuniques[i]<3:
        ord_c.append(i)
    else:
        if i!="city":
            ohe_c.append(i)
```

```python
test=churn.copy()
```

```python
x=test.drop("customerstatus",axis=1)
le=LabelEncoder()
y=test.customerstatus
y=le.fit_transform(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=40,test_size=.20,stratify=y)
```

```python
ct=ColumnTransformer([("HashingEncoder",ce.HashingEncoder(),["city","zipcode"]),("oheencoder",ce.OneHotEncoder(),['offer','multiplelines','internettype','onlinesecurity','onlinebackup','deviceprotectionplan','premiumtechsupport','streamingtv','streamingmovies','streamingmusic','unlimiteddata','contract','paymentmethod']),("ordencoder",ce.OrdinalEncoder(),['gender', 'married', 'phoneservice', 'internetservice', 'paperlessbilling'])],remainder="passthrough")
```

```python
modelpipe=Pipeline([("Transformer",ct),("powertrans",PowerTransformer(standardize=True)),("model",GradientBoostingClassifier(n_estimators=1000,loss="log_loss",learning_rate=0.4,criterion="squared_error",random_state=20,n_iter_no_change=15))])
```

```python
modelpipe.fit(x_train,y_train)
```

```python
y_pred=modelpipe.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
with open("./models/pipemodel.pkl","wb") as fd:
    fd.write(pickle.dumps(modelpipe))
```

```python
with open("./models/pipemodel.pkl","rb") as fd:
    model=pickle.loads(fd.read())
```

```python
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```
